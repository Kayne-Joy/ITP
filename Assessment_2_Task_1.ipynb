*List Products and Words*

```python
# List Products - Retail Company Selected Specialises in Motor Vehicle Products
Products = ["oil",
    "coolant",
    "degreaser",
    "spanner"
]

# List Words
Words = [
    "buy",
    "discount",
    "sale",
    "prices"
]


#Test Lists
print("Product List is:")
print(Products)

print("Word List is:")
print(Words)
```


---
 
*Build Dataframe*


```python
# Import Pandas Dataframe
import pandas as pd

#Create Columns for Dataframe
ad_group_column = []  # Ad group
keywords_column = []  # Keyword column (i.e. "keyword 2a - oil buy" )
keywords_list = []    # Keyword List for Wordcloud (excludes "keyword ... -")

# Initialise dataframe
df = pd.DataFrame({
    "Ad Group" : ad_group_column,
    "Keyword" : keywords_column
})

#Show Dataframe
display(df)
```

---
*Combine Words and Add to Dataframe*

```python
#Combine Words (Test Run)
for product in Products:
  for word in Words:
    Keywords = [product + " " + word, word + " " + product]
    
    #Check Working          !Switchced Off Once Confirmed. Option Avaialble for Future Debugging
    #print (Keywords)   
```

```python
#Add Words to Respective Columns

i = 1                                                         #Define Initial Counter ... Product Number
for product in Products:

  j = "a"                                                     #Define Secondary Counter ... Keyword Letter
  for word in Words:

    Keywords = [product + " " + word, word + " " + product]   #Get KeyWords

    for keyword in Keywords:                                  #Add to Dataframe
      
      keywords_list.append(keyword)                           #Update Keyword List (For WordCloud)

      keyword = "keyword " + str(i) + j + " - " + keyword     #Format keyword                           

      ad_group_column.append(product.capitalize())            #Update Adgroup Column
      keywords_column.append(keyword)                         #Update Keyword Column

      j = chr(ord(j) + 1)                                     #Increase Keyword letter by 1 letter

  i = i + 1                                                   #Increase Product Number by 1

#Check Loop Success
#print(keywords_column)                                       !Switchced Off Once Confirmed. Option Avaialble for Future Debugging
```

```python
#Update Dataframe
df = pd.DataFrame({
    "Ad Group" : ad_group_column,
    "Keyword" : keywords_column
})

#Show Dataframe
display(df)
```


---
*Export as CSV File*


```python
df.to_excel("Question 1 - Marketing Keywords.xlsx")
```


---
*Generate Word Cloud Containing Data*


```python
#Import Required Modules
from wordcloud import WordCloud
import matplotlib.pyplot as plt

#Generate and Format ColorMap
wordcloud = WordCloud(background_color="white", relative_scaling=0.75, colormap ="plasma").generate(" ".join(keywords_list))

#Display Image
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
